#### import things

```python
from matplotlib import pylab as plt
%matplotlib inline
import numpy as np
import math
import scipy
```

#### given function

```python
x = np.arange(1, 15, 0.1)
y = lambda x: np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
plt.plot(x, y(x))
plt.show()
```

#### polynominal pow = 1

```python
b1=([y(1), y(15)])
x1 = np.array([[1, 1], [1, 15]])
```

```python
from scipy import linalg
a1 = scipy.linalg.solve(x1,b1)
```

```python
y1 = lambda x: a1[1]*x + a1[0]
```

```python
plt.plot(x, y1(x))
plt.plot(x, y(x))
plt.show()
```

#### polynomial pow = 2

```python
b2=[y(1), y(15), y(8)]
x2 = np.array([[1, 1, 1], [1, 15, 225], [1, 8, 64]])
```

```python
from scipy import linalg
a2 = scipy.linalg.solve(x2,b2)
```

```python
y2 = lambda x: a2[2]*x**2 + a2[1]*x + a2[0]
plt.plot(x, y2(x))
plt.plot(x, y1(x))
plt.plot(x, y(x))
plt.show()
```

#### polinomial pow = 3

```python
b3=[y(1), y(4), y(10), y(15)]
x3 = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
```
