```python
f = open('sentences.txt')
```

```python
sen = []
for line in f:
    sen.append(line)
```

```python
import re
sentences = []
for sentence in sen:
    sentence=sentence.lower()
    sentence=re.split('[^a-z]', sentence)
    sentence = list(filter(None, sentence))
    sentences.append(sentence)
    print(sentences)
```

```python
myset = set(sentences[0])
for sentence in sentences:
    myset = myset|set(sentence)
```

```python
keys = [i for i in range(len(myset))]
```

```python
mydict = dict(zip(keys, myset))
```

```python
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
c = np.zeros((22, 254))
k=0
for sentence in sentences:
    for key,value in mydict.items():
        if (sentence.count(value) != 0):
            c[k,key]= sentence.count(value)
    k=k+1      
```

```python
from scipy.spatial.distance import cosine
```

```python
dist={}
for i in range(1,len(c)):
    c[i]=c[i].reshape(1,254)
    dist[i]=cosine(c[0],c[i])
print(sorted( ((v,k) for k,v in dist.items())))  
    
```
